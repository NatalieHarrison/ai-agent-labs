### Writing an MCP server that uses the date

```python
params = {"command": "uv", "args":["run", "dates_server.py"]}
async with MCPServerStdio(params=params, client_session_timeout_seconds=30) as server:
  mcp_tools = await server.list_tools()

mcp_tools
```

```python
instructions = "You are a helpful assistant that can use the MCP server that I have created to get the current date."
request = "What is the current date? And give me a fun fact about something in history that happened on this date."
model = "gpt-4o-mini"
```

```python
async with MCPServerStdio(params=params, client_session_timeout_seconds=30) as mcp_server:
  agent = Agent(name="date_manager", instructions=instructions, model=model, mcp_servers=[mcp_server])
  with trace("date_manager"):
    result = await Runner.run(agent, request)
  display(Markdown(result.final_output))


```
